# A closure in simple terms is an inner function that rememebers and has access to variables in the local scope in which it was created even after the outer function has finished executing.

```python
def outer_func():
    message = "Hi"
    #message here is termed as a free variable.
    def inner_func():
        print(message)
    return inner_func()
```

# a closure closes over the free variables from their environment

```python
def outer_func(msg):
    message = msg
    #message here is termed as a free variable.
    def inner_func():
        print(message)
    return inner_func()
```

```python
hi_func = outer_func("Hi")
hello_func = outer_func("Hello")
```

# Practical example

```python
import logging
logging.basicConfig(filename='example.log',level=logging.INFO)

def logger(func):
    def log_func(*args):
        logging.info("Running {} with arguments {}".format(func.__name__,args))
        print(func(*args))
    return log_func

def add(x,y):
    return x+y

def sub(x,y):
    return x - y

```

```python
add_logger = logger(add)
sub_logger = logger(sub)
```

```python
add_logger(3,3)
add_logger(4,5)
```

```python
sub_logger(10,5)
sub_logger(20,5)
```
