# Passing Function as Argument to other Function

```python
def my_map(func, arg_list):
    result = []
    for i in arg_list:
        result.append(func(i))
    return result
```

```python
squares = my_map(square,[1,2,3,4])
```

```python
squares
```

```python
cubes = my_map(cube,[1,2,3,4])
```

```python
cubes
```

```python
def logger(msg):
    def log_message():
        print('Log',msg)
    return log_message
```

```python
log_hello = logger("Hello!!")
```

```python
log_hello
```

```python
log_hello()
```

# Practical Example 

```python
def html_tag(tag):
    def wrap_text(msg):
        print('<{0}>{1}</{0}>'.format(tag,msg))
    return wrap_text
```

```python
print_h1 = html_tag('h1')
```

```python
print_h1
```

```python
print_h1("Test Headline")
```

```python
print_h1("Second Test Headline")
```

```python
print_p = html_tag('p')
```

```python
print_p("Test Paragraph")
```
