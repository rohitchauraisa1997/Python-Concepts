# Map Function

```python
# Direct Method
import math

def area(r):
    return math.pi * (r ** 2)

radii = [1,2,3,4,5,6]
areas = []
for r in radii:
    a = area(r)
    areas.append(a)

print(areas)

print("."*100)
```

```python

# using Map Function

map_op = map(area,radii)
print(map_op)
print(list(map(area,radii)))
```

```python
# Example
# map_func_syntax.png
# Comverting Celsius to Fahrenheit
temps = [("delhi",30),("Berlin",29),("Delhi",40),("Buenos Aires",19)]

def c_to_f(tup):
    return (tup[0], 9/5*tup[1]+32)

converted_list = list(map(c_to_f,temps))
print(converted_list)
```

```python
temps = [("Berlin",29),("Delhi",40),("Buenos Aires",19),("LA",26),("Tokyo",27),("Ney York",28)]
c_too_f = lambda data: (data[0], (9/5)*data[1]+ 32)
# c_too_f = lambda data: print(data[0],"===>",(data[1])+32)

list(map(c_too_f, temps))
```

# Filter Function

```python
import statistics
data = [1,23,23,23,12,32,43,34,54.2,-1.3,-123]
avg = statistics.mean(data)
```

```python
avg
```

```python
filter(lambda x:x>avg, data)
```

```python
list(filter(lambda x:x>avg, data))
```

```python
list(filter(lambda x:x<avg and x>0, data))
```

```python
list(filter(lambda x:x<avg, data))
```

# Remove Missing Data Using Filter

```python
countries = ["India","Brazil","Chile","","Colombia"]
```

```python
# this way we can filter all values that are false
list(filter(None,countries))
```
