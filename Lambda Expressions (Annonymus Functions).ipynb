# Lambda Function with Multiple Inputs 

```python
fullname = lambda fn, ln: fn.strip().title() + " " +ln.strip().title()
```

```python
fullname("Rohit   ","Chaurasia  ")
```

```python
fullname("  rohit   ","chaurasia  ")
```

# Examples

```python
names = ["Rohit","Vatsal","Yogi","Vikash","Pandey","Bhaskar","Sid","Vaibhav","Vedant","Sajjan"]
```

```python
names.sort(key=lambda name:name.split(" ")[-1].lower())
```
